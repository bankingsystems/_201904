<h1> ABBA Simulation</h1>
<p>It shows how to execute MESA version of ABBA. Minimum MESA version is 0.8.5.</p>

```python
from banksim.model import BankSim
from banksim.agent.bank import Bank
from banksim.agent.saver import Saver
from banksim.agent.loan import Loan
```

```python
# Initialization
model = BankSim(initial_saver=10000, 
                  initial_bank=10, 
                  initial_loan=20000, 
                  initial_equity=100,
                  rfree=0.01, # Risk free rate
                  car=0.08 # Minimum regulatory capital adequacy ratio
                 )
```

```python
# Bank status 
for x in [x for x in model.schedule.agents if isinstance(x, Bank)]:
    print("Bank{0:1d} - Reserve: {1:5.0f}, Equity: {2:5.0f}, Deposit: {3:5.0f}, Reserve_ratio: {4:5.3f}".format(x.pos, x.bank_reserves, x.equity, x.bank_deposits, x.reserves_ratio))
```

```python
# Run simulation
[res_bank, res_ibloan] = model.run_model(step_count = 200)
```
